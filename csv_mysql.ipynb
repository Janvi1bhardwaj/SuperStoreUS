```python
import pandas as pd
from sqlalchemy import create_engine
import os

# List of CSV files and their corresponding table names
csv_files = [
    ('customers.csv', 'customers'),
    ('orders.csv', 'orders'),
    ('sellers.csv', 'sellers'),
    ('products.csv', 'products'),
    ('geolocation.csv', 'geolocation'),
    ('payments.csv', 'payments'),
    ('order_items.csv', 'order_items')
]

# Database configuration
database_username = 'root'
database_password = 'gunnu143aju'  # Add your MySQL password here
database_ip       = 'localhost'
database_name     = 'ecommerce'

# Create an engine to connect to the MySQL database
engine = create_engine(f'mysql+pymysql://{database_username}:{database_password}@{database_ip}/{database_name}')

# Folder containing the CSV files
folder_path = 'C:/Users/as/OneDrive/Desktop/dataset'

for csv_file, table_name in csv_files:
    file_path = os.path.join(folder_path, csv_file)
    
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(file_path)
    
    # Replace NaN with None to handle SQL NULL
    df = df.where(pd.notnull(df), None)
    
    # Debugging: Check for NaN values
    print(f"Processing {csv_file}")
    print(f"NaN values before replacement:\n{df.isnull().sum()}\n")

    # Clean column names
    df.columns = [col.replace(' ', '_').replace('-', '_').replace('.', '_') for col in df.columns]

    # Dump the DataFrame to the specified table in the MySQL database
    try:
        df.to_sql(name=table_name, con=engine, if_exists='replace', index=False)
        print(f"Data from {csv_file} has been successfully dumped into the table '{table_name}' in the database '{database_name}'.")
    except Exception as e:
        print(f"Error dumping data from {csv_file} into {table_name}: {e}")

# Close the connection
engine.dispose()
```
