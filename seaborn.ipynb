# seaborn 
scatter plot, customizing with matplotlib, box plot, violin plot, swarm plot,overlaying plot, heatmap, histogram, bar plot, factor plot, density plot, joint distribution plot


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

# line plot

```python
var=[1,2,3,4,5,6,7]
var_1=[2,3,4,1,5,6,7]
x_1=pd.DataFrame({"var":var,"var_1":var_1})
sns.lineplot(x="var",y="var_1",data=x_1)
plt.show()
```

```python
var=[1,2,3,4,5,6,7]
var_1=[2,3,4,1,5,6,7]
sns.lineplot(x=var,y=var_1)
plt.show()
```

```python
y_1=sns.load_dataset("penguins")
y_1
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=y_1,hue="sex",size=10,style="sex",palette="Accent",markers=["<","o"],
             dashes=False,legend="brief")
plt.grid()
plt.title("line plot in seaborn",fontsize=15)
plt.show()
```

```python
sns.set(style="darkgrid")
sns.barplot(x="island",y="bill_length_mm",data=y_1,hue="sex",order=["Biscoe","Dream","Torgersen"],hue_order=["Female","Male"],errorbar=('ci',50),
            n_boot=2,orient='v',palette='dark:g',saturation=0.3,err_kws={'color': 'y'},capsize=1,dodge=True,alpha=0.3,)
plt.show()
```

```python
sns.barplot(x="bill_depth_mm",y="bill_length_mm",data=y_1,orient='v')
plt.show()
```

# histogram plot

```python
y_1=sns.load_dataset("penguins")
sns.displot(x="bill_length_mm",data=y_1,bins=[35,40,45,50,55,60],kde=True,rug=True,color="r",log_scale=False)
plt.show()
```

# scatter plot

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",style="sex",size="sex",sizes=(60,60),palette="gist_ncar",
               alpha=0.6,markers=["*","o"],)
```

# heatmap plot

```python
var=np.linspace(1,10,20).reshape(4,5)
sns.heatmap(var)
plt.show()
```

```python
data=sns.load_dataset("anagrams")
y_2=data.drop(columns=["attnr"],axis=1).head(10)
y_2
```

```python
d={"fontsize":10,"color":"g"}
v=sns.heatmap(y_2,vmin=0,vmax=12,cmap="gist_heat",annot=True,annot_kws=d,linewidth=4,linecolor="y",cbar=False,xticklabels=False,yticklabels=False)
v.set(xlabel="python",ylabel="number")
sns.set(font_scale=6)
plt.show()
```

```python
d=np.linspace(1,10,10).reshape(2,5)
ar=np.array([["a0","a1","a2","a3","a4"],["b0","b1","b2","b3","b4"]])
sns.heatmap(d,vmin=1,vmax=10,cmap="gist_heat",annot=ar,fmt="s")
plt.show()
```
