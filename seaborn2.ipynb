```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
c=sns.load_dataset("tips")
c
```

```python
sns.countplot(x="sex",data=c,hue="smoker",color="g",palette="Accent",saturation=0.1)
plt.show()
```

# violin plot

```python
sns.violinplot(x="day",y="total_bill",data=c,hue="time",linewidth=3,palette="Accent",saturation=0.3,color="g",split=True,inner="quart")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=c,order=["Dinner","Lunch"])
plt.show()
```

```python
sns.violinplot(x=c["total_bill"])
plt.show()
```

# pair plot

```python
sns.pairplot(c,hue="sex")
plt.show()
```

```python
sns.pairplot(c,vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"],palette="BuGn")
plt.show()
```

```python
sns.pairplot(c,hue="sex",hue_order=["Female","Male"],palette="BuGn",x_vars=["total_bill","tip"],kind="reg",diag_kind="hist",markers=["*",">"],)
plt.show()
```

# strip plot

```python
m={"Male":"o","Female":"<"}
sns.stripplot(x="day",y="total_bill",data=c,hue="sex",palette="rocket_r",linewidth=1,edgecolor="y",jitter=5,size=10,marker="*",alpha=0.7)
plt.show()
```

```python
sns.stripplot(x="total_bill",data=c)
plt.show()
```

# box plot

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=c,hue="sex",palette="Dark2",order=["Sat","Sun","Fri","Thur"],showmeans=True,
             meanprops={"marker":"*", "markeredgecolor":"b"},linewidth=3,orient="v",)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="total_bill",data=c)
plt.show()
```

# factor plot -> cat plot

```python
sns.catplot(x="tip",y="size",data=c,palette="Dark2",hue="sex",height=3,kind="point")    # kind (bar,box,scatter,strip,swarm,violin,boxen,point,count)
plt.show()
```

# styling plot

```python
sns.set_style("darkgrid")
sns.barplot(x="day",y="total_bill",data=c)
plt.grid()
plt.show()
```

```python
sns.set_style("white")
sns.set_context("talk",font_scale=1)
plt.figure(figsize=(12,3))
sns.barplot(x="day",y="total_bill",data=c,palette="Accent",hue="day")
sns.despine()
plt.show()
```

# multiple plot

```python
fg=sns.FacetGrid(c,col="sex",hue="day",palette="Accent",)
fg.map(plt.scatter,"total_bill","tip",edgecolor="g").add_legend()
plt.show()
```
