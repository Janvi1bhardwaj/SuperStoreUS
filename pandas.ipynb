 # merge and concat

```python
a1=pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]})
a2=pd.DataFrame({"A":[1,2,3,4,6],"C":[21,22,23,24,25]})
pd.merge(a1,a2,on="A"   #common data only
```

```python
pd.merge(a2,a1,on="A")
```

```python
pd.merge(a2,a1,how="left")  # by default inside 
```

```python
pd.merge(a2,a1,how="right")
```

```python
pd.merge(a2,a1,how="outer")
```

```python
pd.merge(a2,a1,how="outer",indicator=True)
```

```python
a1=pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]})
a2=pd.DataFrame({"A":[1,2,3,4,6],"B":[21,22,23,24,25]})
pd.merge(a1,a2,left_index=True,right_index=True)
```

# pivot function & melt function

```python
d1=pd.DataFrame({"name":["a","b","a","f","r","y","b","e","u"],"mrk_1":[1,2,3,5,4,6,6,7,5],"mrk_2":[2,4,8,9,7,5,4,9,5]})
d1
```

```python
pd.melt(d1)
```

```python
pd.melt(d1,id_vars=["mrk_1"],var_name="janvi",value_name="bhardwaj")
```

```python
d1=pd.DataFrame({"days":[1,2,3,4,5,6],"name":["a","b","c","a","b","c"],"mrk_1":[1,2,3,5,4,7],"mrk_2":[8,2,4,8,9,7]})
d1
```

```python
d1.pivot(index="days",columns="name")
```

```python
d1.pivot(index="days",columns="name",values="mrk_1")
```
