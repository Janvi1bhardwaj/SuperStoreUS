# bar plot

```python
x=[1,2,3,4]
y=[1,2,3,4]
c=["r","b","g","y"]
plt.bar(x,y,color=c)
plt.show()
```

```python
x=["python","c","c++","java"]
y=[98,78,87,88]
c=["r","b","g","y"]
plt.bar(x,y,width=0.3,color=c,align="edge",edgecolor="m",linewidth=4,linestyle=":",alpha=0.5,label="wscube")
plt.legend(loc=2)
plt.xlabel("language",fontsize=10)
plt.ylabel("popluality",fontsize=10)
plt.title("bar graph",fontsize=20)
plt.show()
```

```python
import numpy as np
x=["python","c","c++","java"]
y=[98,78,87,88]
z=[20,30,40,50]

width=0.3
p=np.arange(len(x))
p1=[j+width for j in p]
   
plt.bar(p,y,width,color="r",align="edge",edgecolor="m",linewidth=4,linestyle=":",alpha=0.5,label="wscube")
plt.bar(p1,z,width,color="g",align="edge",edgecolor="m",linewidth=4,linestyle=":",alpha=0.5,label="wscube")
plt.legend(loc=2)
    
plt.xlabel("language",fontsize=10)
plt.ylabel("popluality",fontsize=10)
plt.title("bar graph",fontsize=20)

plt.xticks(p+width/2,x,rotation=20)
    
plt.show()
```

```python
import numpy as np
x=["python","c","c++","java"]
y=[98,78,87,88]
z=[20,30,40,50]

width=0.3
p=np.arange(len(x))
p1=[j+width for j in p]
   
plt.barh(p,y,width,color="r")
plt.barh(p1,z,width,color="g")
plt.legend()
    
plt.xlabel("language",fontsize=10)
plt.ylabel("popluality",fontsize=10)
plt.title("bar graph",fontsize=20)

plt.xticks(p+width/2,x,rotation=20)
    
plt.show()
```

# pie chart

```python
import matplotlib.pyplot as plt
x = [15, 30, 45, 10]
y = ["c", "python", "c++", "java"]
c=["r","b","g","y"]
ex = [0.4, 0.0, 0.0, 0.0]
plt.pie(x, labels=y, explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=1,labeldistance=1,startangle=0,textprops={"fontsize":15},
        counterclock=False,wedgeprops={"linewidth":4,"width":1,"edgecolor":"m"},center=(1,3),rotatelabels=True)
plt.title("pie chart")
plt.legend(loc=3)
plt.show()
```

```python
plt.pie([1])
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [15, 30, 45, 10]
x1=[1,2,3,4]
y = ["c", "python", "c++", "java"]
c=["r","b","g","y"]
plt.pie(x, labels=y, radius=1)
plt.pie(x1,radius=0.5,colors=c)

plt.title("pie chart")
plt.legend(loc=3)
plt.show()
```

```python
x = [15, 30, 45, 10]
y = ["c", "python", "c++", "java"]
c=["r","b","g","y"]
plt.pie(x, labels=y, radius=1)
plt.pie([1],colors="w",radius=0.7)

plt.title("donut chart")
plt.legend(loc=3)
plt.show()
```

```python
x = [15, 30, 45, 10]
y = ["c", "python", "c++", "java"]
c=["r","b","g","y"]
plt.pie(x, labels=y, radius=1)
cr=plt.Circle(xy=(0,0),radius=0.5,facecolor="w")
plt.gca().add_artist(cr)
plt.title("donut chart")
```

# stem plot

```python
x=[1,2,3,4,5,6]
y=[2,4,5,6,2,1]
plt.stem(x,y,linefmt=":",markerfmt="+",bottom=0,basefmt="g",label="python",orientation="horizontal")
plt.legend()
plt.show()
```

# box plot

```python
x=[10,30,50,40,20]
x1=[20,40,60,80,50]
y=[x,x1]
plt.boxplot(y,notch=False,vert=True,widths=0.3,tick_labels=["python","c"],patch_artist=False,showmeans=True,whis=2,sym="g+",boxprops={"color":"r"},
            capprops={"color":"g"},whiskerprops={"color":"b"},flierprops={"markeredgecolor":"y"})

plt.show()
```

# stack plot

```python
x=[1,2,3,4,5,6]
area1=[2,4,5,6,2,1]
area2=[2,4,6,3,1,4]
area3=[6,4,3,2,4,1]
plt.stackplot(x,area1,area2,area3,labels=["area1","area2","area3"],colors=["r","g","y"],baseline="wiggle",)
plt.title("stack plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid()
plt.legend()
plt.show()
```

# step plot

```python
x=[1,2,3,4,5,6]
y=[2,4,5,6,2,1]
plt.step(x,y,color="r",marker="o",ms=10,mfc="g",label="python")
plt.legend()
plt.title("step plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid()
plt.show()
```

# fill_btw plot

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])
plt.plot(x,y,color="r")
plt.fill_between(x,y,color="r",where=(x>=2) & (x<=4),alpha=0.3)
plt.title("fill_btw plot")
plt.xlabel("x-axis")
plt.ylabel("y_axis")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.plot(x,y,color="r")
plt.fill_between(x=[2,4],y1=2,y2=4)
plt.title("fill_btw plot")
plt.xlabel("x-axis")
plt.ylabel("y_axis")
plt.show()
```

# subplot
